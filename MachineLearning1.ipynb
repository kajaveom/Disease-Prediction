```python
data=pd.read_csv('/content/drive/MyDrive/Analytics J Component/encoded.csv') #reading the encoded data file
data1=pd.read_csv('/content/drive/MyDrive/Analytics J Component/UniqueD.csv')   #reading the original file
```

```python
print(data.head(n=5)) #printing the first 5 rows of the emcoded data set.
```

```python
data=data.drop(['X'],axis=1) #dropping the not required column 'X'
```

```python
data=data.drop(['Unnamed: 0'],axis=1) #dropping the not required column 'Unnamed: 0'
```

```python
print(data.head(n=5))  #printing the first fivee records of data after dropping the columns.
```

```python
features=data.iloc[:,1:4] #extracitng the feature set
print(features)
target=data['Disease'] #extracting the target value
print(target)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20) 
```

```python
classifier =DecisionTreeClassifier()     # create a classifier object
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)  #predicting the target value for testing data
```

```python
data_p=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})   #visualing the actaul and predicted value
data_p
```

```python
#testing the predicted value for a record.
test=[[34,36,36]]
test = pd.DataFrame(test,columns=['Symptom_1','Symptom_2','Symptom_3'])
print(test)
predicted=classifier.predict(test)
num=predicted[0]

df2=data1.loc[data1['encodedV'] == num, 'Diseases']
print("The predicted diseases based on the symptoms is: ",df2[0])
```

```python
import math
print("Accuracy:",round(metrics.accuracy_score(y_test, y_pred),4)*100,"%") #printing the accuracy of the model
```

```python
metrics.classification_report(y_test,y_pred)
```

```python
from sklearn.metrics import f1_score, precision_score,recall_score
print("F1 Score : ",f1_score(y_test, y_pred, 
                                           pos_label='positive', average='micro'))
print("Precision Score : ",precision_score(y_test, y_pred,  pos_label='positive',average='micro'))
print("Recall Score : ",recall_score(y_test, y_pred, 
                                           pos_label='positive', average='micro'))
```
